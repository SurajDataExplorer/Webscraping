# Final Code for changing pages and scrape data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Initialize an empty DataFrame to store all data
final = pd.DataFrame()

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.3'}

# Loop through the pages
for j in range(1, 4):  # Scraping only limited pages
    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    webpage = requests.get(url, headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    
    company = soup.find_all('div', class_='companyCardWrapper')
    
    name = []
    ratings = []
    reviews = []
    other_info = []
    
    for i in company:
        name.append(i.find('h2').text.strip())
        ratings.append(i.find('span', class_='companyCardWrapper__companyRatingValue').text.strip())
        reviews.append(i.find('span', class_='companyCardWrapper__ActionCount').text.strip())
        other_info.append(i.find('span', class_='companyCardWrapper__interLinking').text.strip())
    
    d = {'CompanyName': name, 'Ratings': ratings, 'ReviewsCounts': reviews, 'OtherInfo': other_info}
    
    # Create a DataFrame for the current page
    df = pd.DataFrame(d)
    df[['IndustryType', 'Employees', 'CompanyType', 'CompanyAge', 'HQ & Other Locations']] = df['OtherInfo'].str.split(' \| ', expand=True)
    df = df[['CompanyName', 'Ratings', 'ReviewsCounts', 'IndustryType', 'Employees', 'CompanyType', 'CompanyAge', 'HQ & Other Locations']]
    
    # Append the current page data to the final DataFrame
    final = pd.concat([final, df], ignore_index=True)
    

# Display the final DataFrame

final=final[['CompanyName', 'Ratings', 'ReviewsCounts', 'IndustryType', 'Employees']]
#print(final)
```
